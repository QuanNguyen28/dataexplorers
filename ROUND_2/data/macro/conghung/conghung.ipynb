# 1. Cơ cấu ngành của Việt Nam

## 1.1. Tỷ trọng các ngành trong nền kinh tế

```python
import matplotlib.pyplot as plt
import pandas as pd

# Dữ liệu từ file CSV về tỷ trọng các ngành trong nền kinh tế (2020-2024)
data = {
    "Năm": [2020, 2021, 2022, 2023, 2024],
    "Nông, lâm nghiệp và thủy sản": [14.85, 14.00, 11.88, 11.96, 11.86],
    "Công nghiệp và xây dựng": [33.72, 38.00, 38.26, 37.12, 37.64],
    "Dịch vụ": [41.63, 42.00, 41.17, 42.54, 42.22],
    "Thuế sản phẩm trừ trợ cấp sản phẩm": [9.80, 6.00, 8.69, 8.38, 8.28]
}

df = pd.DataFrame(data)

# Đảm bảo năm chỉ có số nguyên
df["Năm"] = df["Năm"].astype(int)

# Tạo subplot với 4 biểu đồ riêng biệt cho từng ngành
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Tỷ trọng các ngành trong nền kinh tế (2020-2024)", fontsize=14)

# Danh sách ngành và vị trí subplot
sectors = list(df.columns[1:])
positions = [(0, 0), (0, 1), (1, 0), (1, 1)]

# Vẽ từng biểu đồ
for sector, pos in zip(sectors, positions):
    ax = axes[pos]
    ax.plot(df["Năm"], df[sector], marker='o', linestyle='-', color='b')
    ax.set_title(sector)
    ax.set_xlabel("Năm")
    ax.set_ylabel("Tỷ trọng (%)")
    ax.set_xticks(df["Năm"])  # Đảm bảo năm hiển thị đúng
    ax.grid(True)

plt.tight_layout(rect=[0, 0, 1, 0.95])  # Điều chỉnh bố cục để tiêu đề không bị che mất
plt.show()
```

## Nhận xét:
- `Nông, lâm nghiệp và thủy sản` có xu hướng giảm dần từ 14.85% (2020) xuống 11.86% (2024). Điều này phản ánh quá trình chuyển dịch cơ cấu kinh tế, khi lao động và nguồn lực chuyển sang công nghiệp và dịch vụ.
- `Công nghiệp và xây dựng` có sự tăng nhẹ từ 33.72% (2020) lên khoảng 38% vào 2021, nhưng sau đó duy trì ở mức ổn định khoảng 37%. Điều này cho thấy ngành này vẫn giữ vai trò quan trọng nhưng không có sự bùng nổ mạnh.
- `Dịch vụ` chiếm tỷ trọng lớn nhất, dao động từ 41.63% đến 42.54%. Điều này cho thấy nền kinh tế đang dần hướng đến mô hình phát triển dựa vào dịch vụ, phù hợp với xu hướng chung của các nước đang phát triển.
- `Thuế sản phẩm trừ trợ cấp sản phẩm` có sự biến động, giảm mạnh vào 2021 (6.00%) nhưng sau đó tăng lên mức 8.28% vào 2024. Điều này có thể phản ánh chính sách thuế hoặc mức độ trợ cấp của nhà nước thay đổi theo từng năm.

## Từ đó ta có thể đánh giá:
- Xu hướng giảm của nông nghiệp và tăng của dịch vụ cho thấy nền kinh tế đang chuyển dịch theo hướng hiện đại hơn.
- Công nghiệp duy trì tỷ trọng ổn định, cho thấy sự đóng góp bền vững nhưng chưa có sự tăng trưởng vượt bậc.
- Biến động thuế có thể liên quan đến điều chỉnh chính sách tài khóa của chính phủ.

## 1.2. Tăng trưởng GDP theo ngành ở Việt Nam

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dữ liệu về Tăng trưởng GDP theo ngành của Việt Nam
data_gdp = {
    "Năm": [2020, 2021, 2022, 2023, 2024],
    "Nông, lâm nghiệp và thủy sản": [2.5, 2.8, 3.0, 3.2, 3.5],
    "Công nghiệp và xây dựng": [7.1, 6.5, 7.0, 6.8, 7.2],
    "Dịch vụ": [6.3, 6.0, 6.8, 7.0, 7.5]
}

# Chuyển đổi thành DataFrame
df_gdp = pd.DataFrame(data_gdp)

# Vẽ biểu đồ Tăng trưởng GDP theo ngành
plt.figure(figsize=(10, 6))
for sector in ["Nông, lâm nghiệp và thủy sản", "Công nghiệp và xây dựng", "Dịch vụ"]:
    plt.plot(df_gdp["Năm"], df_gdp[sector], marker='s', linestyle='-', label=sector)

plt.xlabel("Năm")
plt.ylabel("Tăng trưởng GDP theo ngành (%)")
plt.title("Tăng trưởng GDP theo ngành của Việt Nam (2020-2024)")
plt.legend()
plt.grid(True)
plt.show()
```

- Dịch vụ:

## Đánh giá:

# 2. Sự phát triển của các ngành công nghiệp | công nghệ cao (Mỹ, TQ, VN)

## 2.1. Đầu tư R&D của các quốc gia

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dữ liệu về sự phát triển của ngành công nghiệp công nghệ cao
data = {
    "Năm": [2020, 2021, 2022, 2023, 2024],
    "Mỹ - Đầu tư R&D (% GDP)": [3.45, 3.50, 3.55, 3.60, 3.65],
    "Trung Quốc - Đầu tư R&D (% GDP)": [2.4, 2.5, 2.6, 2.7, 2.8],
    "Việt Nam - Đầu tư R&D (% GDP)": [0.53, 0.55, 0.58, 0.60, 0.63]
}

# Chuyển đổi thành DataFrame
df = pd.DataFrame(data)

# Vẽ biểu đồ đầu tư R&D của các quốc gia
plt.figure(figsize=(10, 6))
for country in df.columns[1:]:
    plt.plot(df["Năm"], df[country], marker='o', linestyle='-', label=country)

plt.xlabel("Năm")
plt.ylabel("% GDP Đầu tư vào R&D")
plt.title("Tăng trưởng Đầu tư R&D của Mỹ, Trung Quốc, Việt Nam (2020-2024)")
plt.legend()
plt.grid(True)
plt.show()
```

## Nhận xét:
- Mỹ:
    - Đầu tư R&D của Mỹ luôn duy trì ở mức cao, từ 3.45% GDP (2020) lên 3.65% GDP (2024).
    - Tăng trưởng ổn định cho thấy Mỹ tiếp tục giữ vị thế là quốc gia dẫn đầu về nghiên cứu và phát triển công nghệ cao.
- Trung Quốc:
    - Đầu tư R&D của Trung Quốc có mức tăng đáng kể, từ 2.4% GDP (2020) lên 2.8% GDP (2024).
    - Xu hướng này cho thấy Trung Quốc đang đẩy mạnh đầu tư vào khoa học và công nghệ, đặc biệt là trong các ngành công nghệ cao như AI, 5G và sản xuất chất bán dẫn.
- Việt Nam:
    - Đầu tư R&D của Việt Nam vẫn ở mức thấp, từ 0.53% GDP (2020) lên 0.63% GDP (2024).
    - Dù có tăng trưởng, nhưng so với Mỹ và Trung Quốc, Việt Nam còn khoảng cách khá xa trong đầu tư cho nghiên cứu và phát triển.

## Kết luận:
- Mỹ vẫn là quốc gia đầu tư R&D cao nhất, đảm bảo vị thế dẫn đầu về công nghệ.
- Trung Quốc có tốc độ tăng trưởng mạnh, thể hiện tham vọng trở thành cường quốc công nghệ.
- Việt Nam đang tăng cường đầu tư R&D nhưng mức chi vẫn còn thấp, cần đẩy mạnh hơn nữa để thúc đẩy ngành công nghiệp công nghệ cao.

## 2.4. Số lượng sản phẩm công nghệ cao của các quốc gia

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dữ liệu về sự phát triển của ngành công nghiệp công nghệ cao
data = {
    "Năm": [2020, 2021, 2022, 2023, 2024],
    "Mỹ - Số lượng sản phẩm công nghệ cao (triệu đơn vị)": [500, 550, 600, 650, 700],
    "Trung Quốc - Số lượng sản phẩm công nghệ cao (triệu đơn vị)": [800, 850, 900, 950, 1000],
    "Việt Nam - Số lượng sản phẩm công nghệ cao (triệu đơn vị)": [100, 120, 140, 160, 180]
}

# Chuyển đổi thành DataFrame
df = pd.DataFrame(data)

# Vẽ biểu đồ Số lượng sản phẩm công nghệ cao
plt.figure(figsize=(10, 6))
for country in ["Mỹ - Số lượng sản phẩm công nghệ cao (triệu đơn vị)", "Trung Quốc - Số lượng sản phẩm công nghệ cao (triệu đơn vị)", "Việt Nam - Số lượng sản phẩm công nghệ cao (triệu đơn vị)"]:
    plt.plot(df["Năm"], df[country], marker='o', linestyle='-', label=country)

plt.xlabel("Năm")
plt.ylabel("Số lượng sản phẩm công nghệ cao (triệu đơn vị)")
plt.title("Số lượng sản phẩm công nghệ cao của Mỹ, Trung Quốc, Việt Nam (2020-2024)")
plt.legend()
plt.grid(True)
plt.show()
```

    - Trung Quốc tiếp tục khẳng định vai trò là công xưởng sản xuất công nghệ cao của thế giới, đặc biệt trong điện thoại thông minh, linh kiện bán dẫn, xe điện và thiết bị viễn thông.

## 2.5. Tỷ lệ lao động trong ngành công nghệ cao của các quốc gia

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dữ liệu về sự phát triển của ngành công nghiệp công nghệ cao
data = {
    "Năm": [2020, 2021, 2022, 2023, 2024],
    "Mỹ - Tỷ lệ lao động công nghệ cao (%)": [10.5, 11.0, 11.8, 12.5, 13.0],
    "Trung Quốc - Tỷ lệ lao động công nghệ cao (%)": [8.2, 8.5, 9.0, 9.6, 10.2],
    "Việt Nam - Tỷ lệ lao động công nghệ cao (%)": [2.5, 2.8, 3.2, 3.7, 4.1]
}

# Chuyển đổi thành DataFrame
df = pd.DataFrame(data)

# Vẽ biểu đồ Tỷ lệ lao động trong ngành công nghệ cao
plt.figure(figsize=(10, 6))
for country in ["Mỹ - Tỷ lệ lao động công nghệ cao (%)", "Trung Quốc - Tỷ lệ lao động công nghệ cao (%)", "Việt Nam - Tỷ lệ lao động công nghệ cao (%)"]:
    plt.plot(df["Năm"], df[country], marker='o', linestyle='-', label=country)

plt.xlabel("Năm")
plt.ylabel("Tỷ lệ lao động trong ngành công nghệ cao (%)")
plt.title("Tỷ lệ lao động trong ngành công nghệ cao của Mỹ, Trung Quốc, Việt Nam (2020-2024)")
plt.legend()
plt.grid(True)
plt.show()
```
