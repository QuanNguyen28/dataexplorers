# Ph√¢n t√≠ch d·ªØ li·ªáu t·ªïng quan v·ªÅ c√°c c√¥ng ty h√†ng ƒë·∫ßu tr√™n th·ªã tr∆∞·ªùng ch·ª©ng kho√°n Vi·ªát Nam

```python
import os
import pandas as pd

# ƒê∆∞·ªùng d·∫´n t·ªõi th∆∞ m·ª•c ch·ª©a c√°c file CSV
folder_path = "."

# Nh√≥m 1: C√°c file ng√¢n h√†ng
bank_files = ['ACB.csv', 'BID.csv', 'CTG.csv', 'HDB.csv', 'LPB.csv', 'MBB.csv', 'SHB.csv',
              'SSB.csv', 'STB.csv', 'TCB.csv', 'TPB.csv', 'VCB.csv', 'VIB.csv', 'VPB.csv']

# Nh√≥m 2: C√°c file phi ng√¢n h√†ng
non_bank_files = ['BCM.csv', 'BVH.csv', 'FPT.csv', 'GAS.csv', 'GVR.csv', 'HPG.csv',
                  'MSN.csv', 'MWG.csv', 'PLX.csv', 'SAB.csv', 'SSI.csv', 'VHM.csv',
                  'VIC.csv', 'VJC.csv', 'VNM.csv', 'VRE.csv']

# H√†m g·ªôp nh√≥m file CSV v√†o m·ªôt DataFrame
def load_and_concat(files, folder_path):
    dfs = []
    for file in files:
        path = os.path.join(folder_path, file)
        df = pd.read_csv(path)
        company_name = file.split('.')[0]
        df.insert(0, 'company', company_name)  # Th√™m c·ªôt company v√†o ƒë·∫ßu
        dfs.append(df)
    return pd.concat(dfs, ignore_index=True)

# T·∫°o hai DataFrame
df1 = load_and_concat(bank_files, folder_path)
df2 = load_and_concat(non_bank_files, folder_path)

# Ki·ªÉm tra s∆° b·ªô
print("üìä df1 - Ng√¢n h√†ng:")
print(df1.head(14))
print("\nüìä df2 - Phi ng√¢n h√†ng:")
print(df2.head(16))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# ·∫®n c√°c c·∫£nh b√°o
warnings.filterwarnings("ignore")

# L·∫•y danh s√°ch c√°c ng√¢n h√†ng
banking_companies = df1['company'].tolist()
no_shareholders = df1['no_shareholders'].tolist()
foreign_percent = df1['foreign_percent'].tolist()
outstanding_share = df1['outstanding_share'].tolist()
issue_share = df1['issue_share'].tolist()
no_employees = df1['no_employees'].tolist()
stock_rating = df1['stock_rating'].tolist()

# Thi·∫øt l·∫≠p bi·ªÉu ƒë·ªì subplot
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle("So s√°nh c√°c ng√¢n h√†ng", fontsize=16)

# Danh s√°ch ti√™u ƒë·ªÅ v√† d·ªØ li·ªáu t∆∞∆°ng ·ª©ng
titles = ["S·ªë l∆∞·ª£ng c·ªï ƒë√¥ng", "T·ª∑ l·ªá s·ªü h·ªØu n∆∞·ªõc ngo√†i (%)", "C·ªï phi·∫øu ƒëang l∆∞u h√†nh", 
          "C·ªï phi·∫øu ƒë√£ ph√°t h√†nh", "T·ªïng s·ªë nh√¢n vi√™n", "X·∫øp h·∫°ng c·ªï phi·∫øu"]
data = [no_shareholders, foreign_percent, outstanding_share, issue_share, no_employees, stock_rating]

y_labels = ["S·ªë l∆∞·ª£ng", "%", "Tri·ªáu c·ªï phi·∫øu", "Tri·ªáu c·ªï phi·∫øu", "S·ªë nh√¢n vi√™n", "X·∫øp h·∫°ng"]

# V·∫Ω t·ª´ng bi·ªÉu ƒë·ªì trong subplot
for i, ax in enumerate(axes.flat):
    sns.barplot(x=banking_companies, y=data[i], ax=ax, palette="Blues_r")
    ax.set_title(titles[i], fontsize=14)
    ax.set_xlabel("Ng√¢n h√†ng")
    ax.set_ylabel(y_labels[i])
    ax.set_xticklabels(banking_companies, rotation=45)

# ƒêi·ªÅu ch·ªânh kho·∫£ng c√°ch gi·ªØa c√°c subplot
plt.tight_layout(rect=[0, 0, 1, 0.96])

# Hi·ªÉn th·ªã bi·ªÉu ƒë·ªì
plt.show()
```
